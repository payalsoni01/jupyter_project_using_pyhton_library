```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("submission.csv")
df.head()
```

```python
df.tail()
```

```python
df.info()
df.describe()
```

```python
plt.figure(figsize=(7,5))
plt.hist(df['diagnosed_diabetes'], bins=30)
plt.title("Distribution of Diagnosed Diabetes Probability")
plt.xlabel("Diabetes Probability")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.kdeplot(df['diagnosed_diabetes'], fill=True)
plt.title("Density Plot of Diabetes Probability")
plt.xlabel("Diabetes Probability")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(df['diagnosed_diabetes'], fill=True)
plt.title("Density Plot of Diabetes Probability")
plt.xlabel("Diabetes Probability")
plt.show()
```

```python
df['diabetes_label'] = df['diagnosed_diabetes'].apply(
    lambda x: 1 if x >= 0.5 else 0
)

df.head()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='diabetes_label', data=df)
plt.title("Count of Diabetic vs Non-Diabetic")
plt.xlabel("Diabetes Label (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```
